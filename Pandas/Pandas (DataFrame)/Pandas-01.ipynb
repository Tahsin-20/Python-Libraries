```python
# Write a function what will return track records of 2 teams against each other
def track_record(team1,team2):
  mask1=(ipl['Team1']==team1) & (ipl['Team2']==team2)
  mask2=(ipl['Team1']==team2) & (ipl['Team2']==team1)

  final= ipl[mask1 | mask2]

  team1_wins = final[final['WinningTeam'] == team1].shape[0]
  team2_wins = final[final['WinningTeam'] == team2].shape[0]
  total_matches = final.shape[0]

  return team1_wins, team2_wins, total_matches

mumbai_wins, chennai_wins, total = track_record('Mumbai Indians','Chennai Super Kings')
print(f"Mumbai Indians Won: {mumbai_wins} matches out of {total}")
print(f"Chennai Super Kings Won: {chennai_wins} matches out of {total}")
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
# Finds out euclidean distance between two points
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5

points_df['distance'] = points_df.apply(euclidean,axis=1) #Mention axis for row
points_df
```
