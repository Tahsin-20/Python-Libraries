### np.sort

Return a sorted copy of an array.

https://numpy.org/doc/stable/reference/generated/numpy.sort.html

```python
#Sorts an array
a = np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)[::-1]
```

### np.append

The numpy.append() appends values along the mentioned axis at the end of the array

https://numpy.org/doc/stable/reference/generated/numpy.append.html

```python
#Appends any number of array
np.append(a,200)
```

```python
np.append(b,np.random.random((b.shape[0],1)),axis=1)
```

### np.concatenate

numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

```python
#Concatenates multiple arrays (Can be used as a replacement of Vertical & Horizontal Stacking)
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)

np.concatenate((c,d),axis=0)
np.concatenate((c,d),axis=1)
```

### np.where

The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
# Find all indices with value greater than 50
np.where(a>50)
```

```python
# Replace all values > 50 with 0 (Condition, Value if true, Value if False)
np.where(a>50,0,a)
```

### np.argmax & np.argmin

The numpy.argmax() function returns indices of the max element of the array in a particular axis.

https://numpy.org/doc/stable/reference/generated/numpy.argmax.html

```python
#Finds the index of largest number in the array
np.argmax(a)
```

```python
np.argmax(b,axis=0) #Columnwise
np.argmax(b,axis=1) #Row-wise
```

```python
# Finds the index of smallest number in the array
np.argmin(a)
```

### np.corrcoef

Return Pearson product-moment correlation coefficients.

https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html

```python
#Finds out correlation between multiple objects
salary = np.array([20000,40000,25000,35000,60000])
experience = np.array([1,3,2,4,2])

np.corrcoef(salary,experience)
```

### Set functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d

```python
#Finds union of multiple arrays
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

np.union1d(m,n)
```

```python
#Finds intersection of multiple arrays
np.intersect1d(m,n)
```

```python
#Finds elements (n,m) which are in n but not in m
np.setdiff1d(n,m)
```

```python
#Gets all the elements except the common (X-OR)
np.setxor1d(m,n)
```
